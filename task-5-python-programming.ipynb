```python
#6 Regex Substitution

import re

n = int(input())
for i in range(n):
    k = input()
    while k.count(' && ')>0:
        k = ' and '.join(re.split(r'\ &&\ ',k))
    while k.count(' || ')>0:
        k = ' or '.join(re.split(r' \|\| ',k))
    print(k)
```

```python
#11 HTML Parser - Part 1

from html.parser import HTMLParser


class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print ('Start :', tag)
        for ele in attrs:
            print ('->', ele[0], '>', ele[1])

    def handle_endtag(self, tag):
        print ('End   :', tag)

    def handle_startendtag(self, tag, attrs):
        print ('Empty :', tag)
        for ele in attrs:
            print ('->', ele[0], '>', ele[1])


parser = MyHTMLParser()
for _ in range(int(input())):
    parser.feed(input())
```

```python
#12 HTML Parser - Part 2

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_comment(self, data):
        if (len(data.split('\n')) != 1):
            print(">>> Multi-line Comment")
        else:
            print(">>> Single-line Comment")
        print(data.replace("\r", "\n"))
    def handle_data(self, data):
        if data.strip():
            print(">>> Data")
            print(data)
  
  
html = ""       
for i in range(int(input())):
    html += input().rstrip()
    html += '\n'
    
parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#13 Detect HTML Tags, Attributes and Attribute Values

from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print(tag)
        for attr, value in attrs:
            print("->", attr, ">", value)

    def handle_startendtag(self, tag, attrs):
        print(tag)
        for attr, value in attrs:
            print("->", attr, ">", value)

html = ''
for _ in range(int(input())):
    html += input().rstrip() + '\n'

parser = MyHTMLParser()
parser.feed(html)
parser.close()
```

```python
#16 Validating Credit Card Numbers

def printRLE(s):
    i = 0
    while (i < len(s) - 1):
        count = 1
        while s[i] == s[i + 1]:
            i += 1
            count += 1
            if i + 1 == len(s):
                break
        i += 1
        if count>=4:
            break
    return count

def check(i):
    k = []
    if (len(i) == 16) and i.isnumeric():
        if str(i[0]) == '4' or str(i[0]) == '5' or str(i[0]) == '6':
                if printRLE(i)<4:
                    print("Valid")
                else:
                    print('Invalid')
        else:
            print('Invalid')
    elif (len(i) == 19) and [str(i[0]) == '4' or str(i[0]) == '5' or str(i[0]) == '6']:
        if (str(i[4])== '-' and str(i[9])== '-' and str(i[14])== '-') :
            if str(i[0]) == '4' or str(i[0]) == '5' or str(i[0]) == '6':
                m = i.replace('-','')
                if m.isnumeric() and printRLE(m)<4:
                    print('Valid')
                else:
                    print('Invalid')
            else:
                print('Invalid')
        else:
            print('Invalid')
    else:
        print('Invalid')


s = []
n = int(input())
for i in range(n):
    s.append(input())
for i in s:
    check(i)
```
