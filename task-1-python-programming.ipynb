```python
#2 Python If-Else

if __name__ == '__main__':
    n = int(input().strip())


    if n%2 == 0:
        if n>=2 and n<=5:
            print('Not Weird')
        elif n>=6 and n<=20:
            print('Weird')
        elif n>20:
            print('Not Weird')
    else:
        print('Weird')
```

```python
#5 Write a function

def is_leap(year):
    leap = False
    if year%100!=0:
        if year%4==0:
            leap =True
        else:
            leap = False
    elif year%100==0:
        if year%400==0:
            leap = True
        else:
            leap = False
    else:
        leap = False
    # Write your logic here
    
    return leap
```
