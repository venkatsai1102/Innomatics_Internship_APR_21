```python
#1 Polar Coordinates

import cmath

number = complex(input())
print(abs(number)) #absolute value
print(cmath.phase(number)) #phase of complex number
```

```python
#2 Find Angle MBC

import math

AB = int(input())
BC = int(input())
angle = math.degrees(math.atan2(AB,BC))
print(str(round(angle))+chr(176))
```
